## challenge

```python
import numpy as np

np.random.seed(42)
n = 150
shop = pd.DataFrame(
    {
        "customer": np.random.choice(
            ["김민준", "이서현", "박지우", "최하은", "정도윤"], n
        ),
        "category": np.random.choice(
            ["의류", "전자기기", "식품", "생활용품", "뷰티"], n
        ),
        "quantity": np.random.randint(1, 6, n),
        "price": np.random.choice([15000, 29000, 45000, 89000, 120000], n),
        "rating": np.random.choice(
            [1, 2, 3, 4, 5], n, p=[0.05, 0.10, 0.20, 0.35, 0.30]
        ),
        "returned": np.random.choice([True, False], n, p=[0.1, 0.9]),
    }
)
shop.loc[np.random.choice(shop.index, 8, replace=False), "rating"] = float("nan")
shop.to_csv("shop_orders.csv", index=False)

df = pd.read_csv("shop_orders.csv")
```

 Q1. 카테고리별 총 매출 내림차순

```python
df.groupby('category')['price'].sum()
```

Q2. 반품률 가장 높은 카테고리

```python
df.groupby('category')['returned'].mean().sort_values(ascending=False).head(1)
```

Q3. rating 결측치를 category별 평균으로 채우기

```python
df.info()
```

```python
df1 = df.copy()
```

```python
df1["rating"] = df1["rating"].fillna(
    df1.groupby("category")["rating"].transform("mean")
)
```

```python
df1.info()
```

Q4. 고객별 총 주문 금액 TOP 3

```python
df1["total_price"] = df["quantity"] * df["price"]
```

```python
df1.groupby("customer")["total_price"].sum().sort_values(ascending=False).nlargest(3)
```
